```python
import folium
import time
from IPython.display import display, clear_output

# Tạo bản đồ trung tâm
m = folium.Map([10.78, 106.70], zoom_start=14)

# Vị trí ban đầu của xe
car_position = [10.78, 106.70]

# Danh sách các điểm đến
destinations = [[10.775, 106.695], [10.785, 106.705], [10.770, 106.688]]

# Thêm marker cho xe
car_marker = folium.Marker(
    car_position,
    popup='Xe',
    icon=folium.Icon(color='blue', icon='car')
).add_to(m)

# Thêm marker cho các điểm đến
for dest in destinations:
    folium.Marker(
        dest,
        popup='Điểm đến',
        icon=folium.Icon(color='green')
    ).add_to(m)

# Di chuyển xe lần lượt đến từng điểm đến
for destination in destinations:
    for step_count in range(10):
        # Nội suy vị trí xe về phía điểm đến
        car_position[0] += (destination[0] - car_position[0]) * 0.3
        car_position[1] += (destination[1] - car_position[1]) * 0.3
        car_marker.location = car_position

        # Cập nhật bản đồ
        clear_output(wait=True)
        display(m)
        time.sleep(0.5)

# Hiển thị bản đồ cuối cùng
display(m)
```
