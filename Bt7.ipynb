```python
import osmnx as ox
import networkx as nx
import folium

# Lấy đồ thị đường phố quận Phú Nhuận, TP.HCM
graph = ox.graph_from_place("Phu Nhuan, HCMC", network_type='drive')

# Danh sách các nút
node_list = list(graph.nodes())
source_node = node_list[0]
target_node = node_list[min(100, len(node_list) - 1)]

# Tìm đường đi ngắn nhất bằng Dijkstra và A*
dijkstra_path = nx.shortest_path(graph, source_node, target_node, 'length')
astar_path = nx.astar_path(graph, source_node, target_node, weight='length')

# Chuyển đổi đường đi thành tọa độ
dijkstra_coords = [(graph.nodes[n]['y'], graph.nodes[n]['x']) for n in dijkstra_path]
astar_coords = [(graph.nodes[n]['y'], graph.nodes[n]['x']) for n in astar_path]

# Tọa độ trung tâm để hiển thị bản đồ
map_center = dijkstra_coords[len(dijkstra_coords) // 2]

# Tạo bản đồ Folium
map_folium = folium.Map(location=map_center, zoom_start=15)

# Vẽ đường đi Dijkstra và A*
folium.PolyLine(dijkstra_coords, color='blue', weight=5, popup='Dijkstra').add_to(map_folium)
folium.PolyLine(astar_coords, color='red', weight=3, popup='A*').add_to(map_folium)

# Đánh dấu điểm bắt đầu và kết thúc
folium.Marker(dijkstra_coords[0], popup='Start', icon=folium.Icon(color='green')).add_to(map_folium)
folium.Marker(dijkstra_coords[-1], popup='End', icon=folium.Icon(color='red')).add_to(map_folium)

# Hiển thị bản đồ (nếu chạy trong Jupyter)
display(map_folium)

# In độ dài các đường đi
print(f"Dijkstra (xanh): {nx.shortest_path_length(graph, source_node, target_node, 'length'):.0f}m")
print(f"A* (đỏ): {nx.astar_path_length(graph, source_node, target_node, weight='length'):.0f}m")
```
