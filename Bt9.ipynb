```python
import folium
import numpy as np
from sklearn.cluster import KMeans

np.random.seed(42)

# Tạo 100 điểm khách hàng ngẫu nhiên xung quanh tọa độ trung tâm (10.78, 106.70)
customers = np.array([
    [10.78 + np.random.normal(0, 0.02), 106.70 + np.random.normal(0, 0.02)]
    for _ in range(100)
])

# Phân cụm khách hàng thành 3 nhóm bằng K-Means
kmeans_model = KMeans(n_clusters=3, random_state=42).fit(customers)

# Tạo bản đồ Folium
m = folium.Map([10.78, 106.70], zoom_start=13)

# Vẽ các điểm khách hàng với màu sắc theo cụm
for idx, (latitude, longitude) in enumerate(customers):
    cluster_color = ['red', 'blue', 'green'][kmeans_model.labels_[idx]]
    folium.CircleMarker([latitude, longitude], radius=2, color=cluster_color).add_to(m)

# Vẽ các điểm trung tâm cụm (địa điểm đặt kho)
for center_lat, center_lon in kmeans_model.cluster_centers_:
    folium.Marker([center_lat, center_lon], popup='Đặt kho', icon=folium.Icon(color='red')).add_to(m)

m
```
