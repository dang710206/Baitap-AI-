```python
import folium
import random
from geopy.distance import geodesic

# Tạo 5 điểm khách hàng với nhiễu ngẫu nhiên
customers = [
    (10.78 + random.uniform(-0.03, 0.03), 106.70 + random.uniform(-0.03, 0.03))
    for _ in range(5)
]

# Tạo 3 điểm trạm với nhiễu ngẫu nhiên
stations = [
    (10.78 + random.uniform(-0.05, 0.05), 106.70 + random.uniform(-0.05, 0.05))
    for _ in range(3)
]

# Bản đồ trung tâm khu vực
map_folium = folium.Map([10.78, 106.70], zoom_start=14)

# Thêm marker cho khách hàng (màu xanh dương)
for idx, customer in enumerate(customers):
    folium.Marker(customer, popup=f'K{idx+1}', icon_color='blue').add_to(map_folium)

# Thêm marker cho trạm (màu xanh lá)
for idx, station in enumerate(stations):
    folium.Marker(station, popup=f'X{idx+1}', icon_color='green').add_to(map_folium)

# Nối mỗi khách hàng với trạm gần nhất
for customer in customers:
    distances = [geodesic(customer, station).km for station in stations]
    nearest_index = distances.index(min(distances))
    folium.PolyLine([customer, stations[nearest_index]], color='red').add_to(map_folium)

# Hiển thị bản đồ
map_folium
```
